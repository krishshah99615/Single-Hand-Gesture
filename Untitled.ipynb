```python
img = cv2.imread(list_of_images[0])

converted = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower = np.array([0, 25, 90], dtype=np.uint8)
upper = np.array([20, 255, 255], dtype=np.uint8)
mask = cv2.inRange(converted, lower, upper)


dilate_kernel = np.ones((1, 1), np.uint8)
it = 2
mask_d = cv2.dilate(mask, dilate_kernel, iterations=it) 


gaussian_kernel = (3, 3)
x_thresh = 100
mask_b = cv2.GaussianBlur(mask_d, gaussian_kernel,
                              x_thresh)
```

```python
plt.imshow(mask_b,cmap='gray')
```
