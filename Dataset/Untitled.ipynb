```python
import cv2
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
list_of_images=glob.glob('pataka/*.jpg')
img = cv2.imread(list_of_images[1])
```

```python
plt.imshow(img)
```

```python
converted = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
lower,upper = np.array([0,25,90],dtype=np.uint8),np.array([20,255,255],dtype=np.uint8)
mask = cv2.inRange(converted,lower,upper)
plt.imshow(mask)
```

```python
mask_d = cv2.dilate(mask,np.ones((1,1),np.uint8),iterations=2)
mask_b = cv2.GaussianBlur(mask_d,(3,3),100)
plt.imshow(mask_b,cmap='gray')
```

```python
def preprocess(name,show=False):
    img = cv2.imread(name)
    converted = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    lower,upper = np.array([0,25,90],dtype=np.uint8),np.array([20,255,255],dtype=np.uint8)
    mask = cv2.inRange(converted,lower,upper)
    mask_d = cv2.dilate(mask,np.ones((1,1),np.uint8),iterations=2)
    mask_b = cv2.GaussianBlur(mask_d,(3,3),100)
    if(show):
        plt.imshow(mask_b,cmap='gray')
        plt.show()
    
    
    
```

```python
for i in list_of_images:
    preprocess(i,True)
```
